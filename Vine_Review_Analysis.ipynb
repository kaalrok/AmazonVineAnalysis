```python
# Import Dependencies
import pandas as pd
```

```python
# Use pgAdmin to export the vine_table as a csv file
# this file has been saved as vine_table.csv in this repository

# Read in the vine_table.csv file
vine_df = pd.read_csv("./Data/vine_table.csv")
vine_df
```

```python
# explore the dataframe using the info method
vine_df.info()
```

## Step 1 ##

Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater tahn 20 to pick reviews that are more likely to be helpful and to avoid having division by zero erros later on.

```python
# filter vine_df for all rowas where total_votes >= 20

vine_df_20 = vine_df.loc[vine_df["total_votes"] >= 20]
vine_df_20

# Note this reduces the total rows from 4,850,360 to 67,855
```

## Step 2 ##

Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater tahn 50%.

```python
# filter vine_df_20 where helpful_votes / total_votes >= 0.50

vine_df_50 = vine_df_20.loc[(vine_df_20["helpful_votes"] / vine_df_20["total_votes"]) >= 0.50]
vine_df_50

# this reduces the number of rows form 67,855 to 61,948
```

## Step 3 ##

Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine =='Y'.

```python
# filter vine_df_50 where vine = 'Y'

paid_reviews_df = vine_df_50.loc[vine_df_50["vine"] == 'Y']
paid_reviews_df

# note there are 334 paid reviews (that meet the filter criteria from steps 1 & 2).
```

## Step 4 ##

```python
# filter vine_df_50 where vine = "N"
unpaid_reviews_df = vine_df_50.loc[vine_df_50["vine"] == 'N']
unpaid_reviews_df
                                   
# note there are 61614 unpaid reviews (that meet the criteria from steps 1 & 2)
```

## Step 5 ##

Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

## Step 5a. Paid Reviews ##

```python
# Determine the count of paid 5 star reviews
paid_5star_reviews = paid_reviews_df.loc[paid_reviews_df["star_rating"] == 5]["star_rating"].count()
paid_5star_reviews
```

```python
# Determine the count of paid reviews
paid_reviews = paid_reviews_df["star_rating"].count()
paid_reviews
```

```python
# Determine the percentage of paid_5star_reviews
# percentage_paid_5star_reviews = paid_5star_reviews / paid_reviews
percentage_paid_5star_reviews = round((paid_5star_reviews / paid_reviews) *100, 2)
percentage_paid_5star_reviews
```

## Step 5b Unpaid Reviews ##

```python
# Determine the count of unpaid 5 star reviews
unpaid_5star_reviews = unpaid_reviews_df.loc[unpaid_reviews_df["star_rating"] == 5]["star_rating"].count()
unpaid_5star_reviews
```

```python
# Determine the count of unpaid reviews
unpaid_reviews = unpaid_reviews_df["star_rating"].count()
unpaid_reviews
```

```python
# Determine the percentage of unpaid_5star_reviews
# percentage_paid_5star_reviews = paid_5star_reviews / paid_reviews
percentage_unpaid_5star_reviews = round((unpaid_5star_reviews / unpaid_reviews) *100, 2)
percentage_unpaid_5star_reviews
```
